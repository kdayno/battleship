```python
import random
from IPython.display import clear_output


def CreateGameBoard():

    game_board = []

    for num in range(7):
        row = [' ', ' ', ' ', '|'] * 7
        game_board.append(row)

    return game_board

def DisplayGameBoard(game_board, board_type):
    
    if board_type == 'Bot':
        print("\n         Bot's Game Board:\n")
    else:   
        print("\n         Your Game Board:\n")
    
    border = '   ____________________________'
    row_num = 1

    print('    A   B   C   D   E   F   G')
    print(border)

    for row in game_board:
        
        print(f'{row_num} ' + '|' + ''.join(row))
        print(border)

        row_num += 1

def PositionCheck(ship_size, column, row, orientation, game_board):

    column_choice = column_index_lookup[column]
    row_choice = row_index_lookup[row]
    
    if orientation == 'Right':

        for num in range(ship_size):

            return False if game_board[row_choice][column_choice] != ' ' else True
                
            column_choice += 4 
    
    elif orientation == 'Left':

        for num in range(ship_size):
            
            return False if game_board[row_choice][column_choice] != ' ' else True

            column_choice -= 4

    elif orientation == 'Up':

        for num in range(ship_size):
            
            return False if game_board[row_choice][column_choice] != ' ' else True
            
            row_choice -= 1

    elif orientation == 'Down':

        for num in range(ship_size):
            
            return False if game_board[row_choice][column_choice] != ' ' else True

            row_choice += 1

def ShipSizeCheck(ship_size, column, row, orientation):
    
    column_choice = column_index_lookup[column]
    row_choice = row_index_lookup[row]

    if orientation == 'Right':
        
        ship_size = column_choice + (ship_size * 4)

        return False if ship_size > 29 else True
    
    elif orientation == 'Left':

        ship_size = column_choice - (ship_size * 4)
        
        return False if ship_size < -3 else True

    elif orientation == 'Up':

        ship_size = row_choice - ship_size

        return False if ship_size < -1 else True

    elif orientation == 'Down':

        ship_size = row_choice + ship_size

        return False if ship_size > 7 else True

def PlaceShip(ship_size, column, row, orientation, game_board):

    column_choice = column_index_lookup[column]
    row_choice = row_index_lookup[row]

    game_board[row_choice][column_choice] = 'O'

    if orientation == 'Right':

        for num in range(ship_size-1):

            column_choice += 4 

            game_board[row_choice][column_choice] = 'O'

    elif orientation == 'Left':

        for num in range(ship_size-1):
            
            column_choice -= 4

            game_board[row_choice][column_choice] = 'O'

    elif orientation == 'Up':

        for num in range(ship_size-1):
            
            row_choice -= 1

            game_board[row_choice][column_choice] = 'O'

    elif orientation == 'Down':

        for num in range(ship_size-1):
            
            row_choice += 1

            game_board[row_choice][column_choice] = 'O'
            
    return game_board

def CreateShipMapping(game_board):

    mapping = []

    for row_index, row in enumerate(game_board):
            
        for space_index, space in enumerate(row):
            
            if space == 'O':

                coordinates = (row_index, space_index)

                mapping.append(coordinates)

    return mapping

def ResetBoard(game_board):

    for row_index, row in enumerate(game_board):
            
        for space_index, space in enumerate(row):
            
            if space == 'O' or space == 'X':

                game_board[row_index][space_index] = ' '

    return game_board

def CheckIfShipExists(x, y, mapping):

    return True if (x,y) in mapping else False

def PlaceMarker(coordinates, game_board, marker):

    row, column = coordinates
    
    game_board[row][column] = marker

def BotShipPlacement():

    column_choices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    row_choices = ['1', '2', '3', '4', '5', '6', '7']
    orientation_choices = ['Up', 'Down', 'Left', 'Right']
    ship_size_choices = [2, 3, 4, 5]

    random_ship_size = random.sample(ship_size_choices, 1)
    random_ship_size = random_ship_size[0]
    
    random_column = random.sample(column_choices, 1)
    random_column = random_column[0]

    random_row = random.sample(row_choices, 1)
    random_row = random_row[0]

    random_orientation = random.sample(orientation_choices, 1)
    random_orientation = random_orientation[0]

    return random_ship_size, random_column, random_row, random_orientation

def BotTargetChoice():

    column_choices = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
    row_choices = ['1', '2', '3', '4', '5', '6', '7']

    random_column = random.sample(column_choices, 1)
    random_column = random_column[0]

    random_row = random.sample(row_choices, 1)
    random_row = random_row[0]

    return random_column, random_row

def GameStatusCheck(player_mapping, bot_mapping):

    if bot_mapping == []:

        return (True, 'Player')
    
    elif player_mapping == []:

        return (True, 'Bot') 

    else:
        
        return (False, None)
```

```python
### BATTLESHIP GAME ###

def BattleShipGame(): #
    
    column_index_lookup = {'A':1, 'B':5, 'C':9, 'D':13, 'E':17, 'F':21, 'G':25}
    row_index_lookup = {'1':0, '2':1, '3':2, '4':3, '5':4, '6':5, '7':6}

    player_game_board = CreateGameBoard()
    bot_game_board = CreateGameBoard()

    print("\n> > > > > B A T T L E S H I P > > > > >")
    DisplayGameBoard(bot_game_board, 'Bot')
    print("\n")
    DisplayGameBoard(player_game_board, 'Player')

    input("Are you ready to play?")

    print("The Bot is placing their ships...")

    ### BOT PLACES SHIPS ###

    bot_ship_counter = 0

    while bot_ship_counter < 3: 

        ship_size, column, row, orientation = BotShipPlacement()

        if ShipSizeCheck(ship_size, column, row, orientation) is True:

            if PositionCheck(ship_size, column, row, orientation, bot_game_board) is True:
                
                PlaceShip(ship_size, column, row, orientation, bot_game_board)
                
                bot_ship_counter += 1
                

    print("The Bot placed their ships.\n")

    clear_output()

    input("Your turn, are your ready to place your ships?")

    DisplayGameBoard(player_game_board, 'Player')

    ### PLAYER PLACES SHIPS ###

    player_ship_counter = 0

    while player_ship_counter < 3:

        ship_size_choice = int(input("What size ship? (2-5)"))

        column_choice = input("Which column do you want to place your first ship?")

        row_choice = input("Which row?")

        orientation_choice = input("Which position?")

        if ShipSizeCheck(ship_size_choice, column_choice, row_choice, orientation_choice) is True:

            if PositionCheck(ship_size_choice, column_choice, row_choice, orientation_choice, player_game_board) is True:
                
                PlaceShip(ship_size_choice, column_choice, row_choice, orientation_choice, player_game_board)

                print("Your ship has been placed.")
            
            else:
                
                print("That position is occupied.")
        else:

            print("Your ship won't fit there.")

        player_ship_counter += 1

        clear_output()

        DisplayGameBoard(player_game_board, 'Player')

    clear_output()

    bot_ship_mapping = CreateShipMapping(bot_game_board)
    player_ship_mapping = CreateShipMapping(player_game_board)

    ResetBoard(bot_game_board)

    DisplayGameBoard(bot_game_board, 'Bot')
    print('\n')
    DisplayGameBoard(player_game_board, 'Player')

    print("Time to Play!")

    clear_output()

    game_check = False

    while game_check is False:

        DisplayGameBoard(bot_game_board, 'Bot')
        DisplayGameBoard(player_game_board, 'Player')

        column = input("Which column?")
        row = input("Which row?")

        player_column_choice = column_index_lookup[column]
        player_row_choice = row_index_lookup[row]

        clear_output()

        if (player_row_choice, player_column_choice) in bot_ship_mapping:
            
            PlaceMarker((player_row_choice, player_column_choice), bot_game_board, 'X')

            bot_ship_mapping.remove((player_row_choice, player_column_choice))

            print(f"\nPlayer targeted: {column}{row}. Player HIT!")

        
        else:

            PlaceMarker((player_row_choice, player_column_choice), bot_game_board, '-')

            print(f"\nPlayer targeted: {column}{row}. Player MISSED.")

        bot_column, bot_row = BotTargetChoice()

        bot_column_choice = column_index_lookup[bot_column]
        bot_row_choice = row_index_lookup[bot_row]

        if (bot_row_choice, bot_column_choice) in player_ship_mapping:

            PlaceMarker((bot_row_choice, bot_column_choice), player_game_board, 'X')

            player_ship_mapping.remove((bot_row_choice, bot_column_choice))

            print(f"\nBot targeted: {bot_column}{bot_row}. Bot HIT!")

        else:

            PlaceMarker((bot_row_choice, bot_column_choice), player_game_board, '-')

            print(f"\nBot targeted: {bot_column}{bot_row}. Bot MISSED.")

        (game_check, winner) = GameStatusCheck(player_ship_mapping, bot_ship_mapping)

    clear_output()

    print(f"\nGame Over! {winner} Wins!")

    DisplayGameBoard(bot_game_board, 'Bot')
    DisplayGameBoard(player_game_board, 'Player')
```
